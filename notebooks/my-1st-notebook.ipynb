```python
import numpy as np
import matplotlib.pyplot as plt

# Create a NumPy array with shape (10, 3) filled with random integers
random_array = np.random.randint(1, 100, size=(10, 3))

# Print the array
print("Random Array:")
print(random_array)

# Create a scatter plot
fig, ax = plt.subplots()
ax.scatter(random_array[:, 0], random_array[:, 1], c=random_array[:, 2], cmap='viridis', s=100)

# Add labels and title
ax.set_xlabel('Column 1')
ax.set_ylabel('Column 2')
ax.set_title('Scatter Plot of Random Array')

# Display the plot
plt.show()
```
